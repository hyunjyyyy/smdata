```python
from IPython.display import display
import pandas as pd
import numpy as np

# ------------------------------
# 1) í‚¤/íƒ€ì… ì •ë¦¬
# ------------------------------
KEYS = ['í”„ë¡œê·¸ë¨ID', 'í”„ë¡œê·¸ë¨ì„¸ë¶€ID']
for c in KEYS:
    df_program[c] = pd.to_numeric(df_program[c], errors='coerce').astype('Int64')
    df_student[c] = pd.to_numeric(df_student[c], errors='coerce').astype('Int64')

# ------------------------------
# 2) ì§‘ê³„ìš© ì»¬ëŸ¼/ë§ˆìŠ¤í‚¹
# ------------------------------
is_completed = df_student['ì´ìˆ˜ì—¬ë¶€'].astype(str).str.strip().eq('ì´ìˆ˜')  # 'ë¯¸ì´ìˆ˜' ì œì™¸

score_cols = [c for c in df_student.columns if c.startswith('í•µì‹¬ì—­ëŸ‰ì ìˆ˜')]
agg_dict = {
    'ì´ìˆ˜ìID': pd.Series.nunique,                    # ì°¸ì—¬ììˆ˜: ìœ ë‹ˆí¬ ì¸ì›
    'ì´ìˆ˜ì—¬ë¶€': lambda s: (s.astype(str).str.strip().eq('ì´ìˆ˜')).sum(),  # ì´ìˆ˜ììˆ˜
    'ë§Œì¡±ë„í‰ê· ': 'mean'
}
for c in score_cols:
    agg_dict[c] = 'mean'

# ------------------------------
# 3) í”„ë¡œê·¸ë¨ë³„ ì§‘ê³„ ê³„ì‚°(df_prog_stats)
# ------------------------------
df_prog_stats = (
    df_student
    .groupby(KEYS, dropna=False)
    .agg(agg_dict)
    .reset_index()
    .rename(columns={
        'ì´ìˆ˜ìID': 'ì°¸ì—¬ììˆ˜_calc',
        'ì´ìˆ˜ì—¬ë¶€': 'ì´ìˆ˜ììˆ˜_calc',
        'ë§Œì¡±ë„í‰ê· ': 'ë§Œì¡±ë„í‰ê· _calc',
        **{c: f'{c}_calc' for c in score_cols}
    })
)

df_prog_stats['ì´ìˆ˜ìœ¨(%)_calc'] = np.where(
    df_prog_stats['ì°¸ì—¬ììˆ˜_calc'].fillna(0) > 0,
    (df_prog_stats['ì´ìˆ˜ììˆ˜_calc'] / df_prog_stats['ì°¸ì—¬ììˆ˜_calc'] * 100).round(1),
    np.nan
)

# ------------------------------
# 4) í”„ë¡œê·¸ë¨ ì›ë³¸ê³¼ ë³‘í•©(ì›ë³¸ ì°¸ì—¬/ì´ìˆ˜ ë³´ì¡´)
# ------------------------------
df_merged = df_program.merge(df_prog_stats, on=KEYS, how='left')

# ì •ë¦¬: calc ìˆ˜ì¹˜í˜• ìºìŠ¤íŒ…/ë°˜ì˜¬ë¦¼
for c in ['ì°¸ì—¬ììˆ˜_calc', 'ì´ìˆ˜ììˆ˜_calc']:
    if c in df_merged.columns:
        df_merged[c] = df_merged[c].fillna(0).astype(int)
if 'ë§Œì¡±ë„í‰ê· _calc' in df_merged.columns:
    df_merged['ë§Œì¡±ë„í‰ê· _calc'] = df_merged['ë§Œì¡±ë„í‰ê· _calc'].round(2)
for c in [f'{s}_calc' for s in score_cols]:
    if c in df_merged.columns:
        df_merged[c] = df_merged[c].round(2)

# ------------------------------
# 5) ë³´ê¸° ì¢‹ê²Œ ì¶œë ¥(ì›ë³¸ vs ì¬ê³„ì‚° ë‚˜ë€íˆ)
# ------------------------------
cols_basic = [c for c in [
    'í”„ë¡œê·¸ë¨ID','í”„ë¡œê·¸ë¨ì„¸ë¶€ID','í”„ë¡œê·¸ë¨ëª…','ì£¼ê´€ë¶€ì„œ(ìƒìœ„)','ì£¼ê´€ë¶€ì„œ(í•˜ìœ„)','ì‹œì‘ì¼','ì¢…ë£Œì¼'
] if c in df_merged.columns]

cols_compare = [c for c in [
    'ì°¸ì—¬ììˆ˜', 'ì´ìˆ˜ììˆ˜',           # ì›ë³¸(ì‹œìŠ¤í…œ) ê°’
    'ì°¸ì—¬ììˆ˜_calc','ì´ìˆ˜ììˆ˜_calc','ì´ìˆ˜ìœ¨(%)_calc','ë§Œì¡±ë„í‰ê· _calc'
] if c in df_merged.columns]

print("ğŸ“Œ í”„ë¡œê·¸ë¨ë³„ ì›ë³¸ vs ì¬ê³„ì‚° ì§€í‘œ (ìƒìœ„ 10)")
display(df_merged[cols_basic + cols_compare].head(10))

print("\nğŸ† ì´ìˆ˜ììˆ˜_calc TOP 20")
display(
    df_merged.sort_values('ì´ìˆ˜ììˆ˜_calc', ascending=False)[cols_basic + cols_compare].head(20)
)
```
